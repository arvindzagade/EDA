```python
# Step 3 --> Relationship Analysis
```

```python
#Checking for outliers ---> Currently we dont have outliers 
```

```python
# Correlation Matrix

corelation = student.corr() 
```

```python
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

```python
# Pairplot

sns.pairplot(student)
```

```python
# Scatter plot ---> shows relationship between two numerical variables

sns.relplot(x='math score',y='reading score',hue='gender',data=student)
```

```python
sns.relplot(x='math score',y='reading score',hue='lunch',data=student)
```

```python
sns.relplot(x='math score',y='reading score',hue='test preparation course',data=student)
```

```python
# histogram ---> grahical display of data using bars of diff height
# it displays shape,spread and continious sample data


sns.distplot(student['math score'],bins=10)
```

```python
sns.distplot(student['reading score'],bins=10)
```

```python
sns.distplot(student['writing score'],bins=10)
```
